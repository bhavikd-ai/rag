```python
# PREPARING DATA FOR EVALUATION

from datasets import Dataset

data = {
    "question": [
        "What is RAG?"
    ],
    "answer": [
        "RAG stands for Retrieval Augmented Generation."
    ],
    "contexts": [[
        "RAG stands for Retrieval Augmented Generation. It combines retrieval and generation."
    ]],
    "ground_truth": [
        "RAG means Retrieval Augmented Generation."
    ]
}

dataset = Dataset.from_dict(data)
```

```python
# INITIATE SENTENCE TRANSFORMER EMBEDDING MODEL

from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# CHECKING THE AVAILABLE GEMINI MODELS

import google.generativeai as genai
import os

for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
# INITIATE GEMINI MODEL

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",
    temperature=0
)
```

```python
# EVALUATING RAG ON FAITHFULNESS, RELEVENCE, PRECISION & RECALL

from ragas import evaluate

from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_precision,
    context_recall
)

result = evaluate(
    dataset=dataset,
    metrics=[
        faithfulness,
        answer_relevancy,
        context_precision,
        context_recall
    ],
    llm=llm,
    embeddings=embeddings
)

print(result)
```
